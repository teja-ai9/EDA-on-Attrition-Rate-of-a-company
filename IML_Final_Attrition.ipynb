**Findings from the Univariant Analysis:**
**Based on Age:**
Range of age is between 18 years to 60 years: This company has employees across the spectrum from recent graduates to the retirement age.
Since the maximum age bracket with highest number of employees are between 30 -36, we can safely assume that the company prefers employees who have considerable work experience.

**Based on Attrition:**
As a general rule, employee retention rates of 90 percent or higher are considered good and a company should aim for a turnover rate of 10% or less.
Since the attrition rate in our data is greater than 10% (16.12%), we can infer that the company is facing difficulties with employee retention.

**Based on Business Travel:****
Since 70% of employees travel rarely, the business travel does not seem to be an integral job-deliverable for majority of the employees.

**Based on Department:**
Since 65% of the employees are in Research and development, we can assume that the company activities include innovation and introducing new products and services. Ostensibly, the goal of the company would be to take new products and services to market and add to the company's bottom line.

**Based on Education Field:**
Since 41% and 31% of the employees are from the Life sciences and medical backgrounds respectively, we can say that the company is majorly based out of bio-technology and R&D. This further strengthens our earlier propositions.

**Based on Environment Satisfaction**
Since more than 60% of the employees have voted for the top 2 categories in environment satisfaction, we can say that the majority of employees feel comfortable working in this organization. However, 40% of the employees seem to need improvement in their working environment. From the above we can say that the organization needs to work on their "hygiene" conditions for employee motivation("Herzberg two factor theory"). 
So, the company seems to have more scope to improve the working conditions, physical workspace and OSH compliance.

**Based on Gender**
Since there are 60% of the male employees and 40% of the female employees, the company would need to strive for equal participation of workforce with respect to gender. 
Taking a broader perspective, this finding gives credence to the social fact that there is a lower proportion of women workforce in Research and development profiles. This situation needs to be improved to achieve gender parity in workforce. 

**Based on Job Involvement**
More than 80% of the employees are in the middle of the cluster with respect to job involvement metrics. This might also be due to a "neutral response bias"(where people choose whatever the middle option is).

**Based on Job Level**
Senior level employees are less in number(4.7%). The junior level employees (Levels 1 and 2) are almost 73.2%, which represents the majority of workforce. So, the company seems to have a typical 'hierarchical' organisational structure.
There could be better levels of authority and responsibility. But this pyramidal chain of command can slow down 'innovation' - which is quintessential for an R&D company.

**Based on Job Satisfaction**
Only 30% of employees in the organization have given the highest rating for job satisfaction. This shows that there is still a lot of room for improvement.
The company can focus on the "Motivator" factors for Job Satisfaction ("Herzberg two factor theory") like:
-Performance and achievement,
-Recognition and personal growth,
-Responsibility and opportunity for advancement, 
-Job status and work itself.

**Based on Marital Status**
As 45% of the employees in the organization are married, the company appears to give a stable and well-paying job profile to the employees. Marital status is also an important indicator of organisational commitment. Generally, married people are more committed to their organisation than unmarried people.

**Based on Monthly Income**
The monthly income appears to be 'positively skewed' since a majority of employees fall in the normal or lower-income earning group rather than a few high earning income groups. This also clearly reiterated by the high number of outliers and the inequality in income distribution.

**Based on Over Time**
Since there are 28% of the employees who are doing overtime, there is a considerable organisational bias towards it. This shows a more flexible workforce, increased earning for employees and mutual employer benefit. However, it could result in inefficiency and it becomes necessary to compensate by providing overtime pay.

**Based on Total Working Years**
Maximum number of employees have a total working experience of 10 years (202 in number). As corroborated with the Age column, many senior professionals are preferring to stay with the organization.

**Based on WorkLifeBalance**
60% of employees have given second best rating. This could again hint at a 'neutral response bias' and it involves a certain subjectivity of interpreting a correct work-life balance. Nonetheless, there appears to be a lot of scope to enhance the work life balance of the employees.

**Based on Years At Company**
The dataset reveals that a majority of the employees (75%) have a work experience of less than 9 years at the company. There appears to be a tapering down of career progression above 10 years. If the company can positively tackle this, it can greatly benefit the organisation.
